
# PRE-PROCESSING

```python
frame.round(decimals=3)
```

```python
for column in frame.columns:
    if (column != 'target'):
        mean = frame[column].mean()
        stdev = frame[column].std()
        frame[column] = (frame[column]-mean)/stdev
```

```python
col = frame.columns[4:10]
col
```

```python
for column in col:
    Q3 = frame[column].quantile(0.75)
    Q1 = frame[column].quantile(0.25)
    IQR = Q3-Q1
    a = (frame[column]<(Q1-1.5*IQR)).sum()
    b = (frame[column]>(Q3+1.5*IQR)).sum()
    print('outliers of {} column is {}'.format(column,a+b))
    
    frame[column] = np.where(frame[column]<(Q1-1.5*IQR),(Q1-1.5*IQR),np.where(frame[column]>(Q3+1.5*IQR),Q3+1.5*IQR,frame[column]))
```
